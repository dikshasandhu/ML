```python
def gradient(x,y,it):
    m=b=0
    rate=0.01
    n=len(x)
    for i in range(it): 
        ypr=m*x +b
        md=(-2/n)*sum(x*(y-ypr))
        bd=(-2/n)*sum(y-ypr)
        m=m-rate*md
        b=b-rate*bd
        print(m,b)
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,0])
gradient(x,y,4)
```
