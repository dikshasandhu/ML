```python
import cv2
img=cv2.imread('cat.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.imshow(img)
```

```python
flat_image=img.reshape((-1,3))
val=np.float32(flat_image)
```

```python
criteria=(cv2.TERM_CRITERIA_EPS +  cv2.TERM_CRITERIA_MAX_ITER,100,0.85)
```

```python
retval, labels, centers = cv2.kmeans(val, 15, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers=np.uint8(centers)

segmented_data=centers[labels.flatten()]
```

```python
k1=segmented_data.reshape(img.shape)
```

```python
plt.imshow(k1)
```

```python
cv2.imwrite('15.jpg',k1)
```
