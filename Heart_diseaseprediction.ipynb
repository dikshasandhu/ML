```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
chest_p=LabelEncoder()
exercise_a=LabelEncoder()
df['chest_pain']=chest_p.fit_transform(df.ChestPainType)
df['exercise_angina']=exercise_a.fit_transform(df.ExerciseAngina)
df.head()
```

```python
df=df.drop(['ChestPainType','RestingBP','RestingECG','ExerciseAngina','ST_Slope','Sex'],axis='columns')
```

```python
x=df.drop('HeartDisease',axis='columns')
y=df.HeartDisease
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10)
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
b=BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
b.fit(xtrain,ytrain)
```

```python
b.score(xtest,ytest)
```
