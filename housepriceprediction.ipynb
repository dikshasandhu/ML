```python
import pandas as pd
d=pd.read_csv('town.csv')
d.head()
```

```python
dummie=pd.get_dummies(d.town)
```

```python
 merge=pd.concat([d,dummie],axis='columns')
```

```python
merge
```

```python
 final=merge.drop(['town','m'],axis='columns')
```

```python
final
```

```python
final=final.dropna(axis=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(final[['area','r','w']],final.price)
```

```python
import numpy as np
inp=np.array([2800,1,0])
t=inp.reshape(1,-1)
pred=model.predict(t)
pred
```
