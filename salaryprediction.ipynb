```python
import pandas as pd
d=pd.read_csv('decision.csv')
d.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
lj=LabelEncoder()
ld=LabelEncoder()
d['co']=lc.fit_transform(d.company)
d['jo']=lc.fit_transform(d.job)
d['de']=lc.fit_transform(d.degree)
d.head()
d=d.drop(['company','job','degree'],axis='columns')
```

```python
d=d.dropna()
d
```

```python
from sklearn import tree
m=tree.DecisionTreeClassifier()
x=d[['co','jo','de']]
y=d[['salary']]
m.fit(x,y)
#m.predict([2,2,0])
m.score(x,y)
```
